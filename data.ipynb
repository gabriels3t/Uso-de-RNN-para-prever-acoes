```python
import yfinance as yf
import numpy as np
```

# Pegando dados e dividindo entre feature e target

```python
# Definir os ativos e o perÃ­odo
ativos4 = ['SUZB3.SA', 'EMBR3.SA', 'PETR4.SA', 'EGIE3.SA', 'BBSE3.SA', 'ITSA4.SA', 'ABEV3.SA', 'WEGE3.SA', 'FLRY3.SA', 'TAEE4.SA', 'USDBRL=X','^BVSP']
data_inicial = '2017-01-01'
#data_inicial = '2024-01-01'
data_final = '2025-12-30'

# Obtendo os dados diretamente com o yfinance
tabela_cotacoes = yf.download(ativos4, start=data_inicial, end=data_final)

# Removendo os NA
cotacao_sem_nulos = tabela_cotacoes.dropna()

cotacao_sem_nulos.head()
```

```python
cotacao_feature = cotacao_sem_nulos  # Features
cotacao_target = cotacao_sem_nulos['Close']  # Target
```

```python
cotacao_target.head()
```

# Agrupando features em 10 dias

```python
# Defina o tamanho da janela
window_size = 10

# Crie as janelas 
#cotacao_feature_grouped = [cotacao_feature.iloc[i:i + window_size] for i in range(0, len(cotacao_feature) - window_size)]
cotacao_feature_grouped = [cotacao_feature.iloc[i:i + window_size] for i in range(0, len(cotacao_feature), window_size)]

# Verifique o nÃºmero de janelas criadas e o tamanho das janelas
print(f"Quantidade de janelas criadas: {len(cotacao_feature_grouped)}")
print(f"Formato da primeira janela: {cotacao_feature_grouped[0].shape}")  # Esperado: (10, 60)

cotacao_feature_grouped[0]
```

O objetivo Ã© prever o valor da aÃ§Ã£o apos 10 dias de dados temporais, com isso os dados de target sÃ£o  separados em 10 em 10 dias, diminuindo o seu tamanho para 189

```python
acao= [
    "ABEV3.SA", "BBSE3.SA", "EGIE3.SA", "EMBR3.SA", "FLRY3.SA", 
    "ITSA4.SA", "PETR4.SA", "SUZB3.SA", "TAEE4.SA",
    "WEGE3.SA"
]
# Ajuste o target para garantir que ele tenha o mesmo nÃºmero de amostras que as features
# O target serÃ¡ o 'Adj Close' do dia seguinte apÃ³s cada janela de 10 dias
cotacao_target_intervalo = cotacao_target.iloc[window_size::window_size][acao]


print(f"Novo shape do target: {cotacao_target_intervalo.shape}")
cotacao_target_intervalo
```

## Utilizando o indice MACD

```python
# FunÃ§Ã£o para calcular MACD e Signal para os ativos em cada janela
def calcular_macd_em_janelas(cotacao_feature_grouped, ativos):
    # Cria uma lista para armazenar as novas janelas com MACD e Signal
    cotacao_feature_with_macd = []

    # Iterar sobre cada janela
    for janela in cotacao_feature_grouped:
        # Iterando sobre cada ativo para calcular o MACD e a linha de sinal
        for ativo in ativos:
            # Seleciona os dados de fechamento do ativo na janela e remove valores nulos
            cotacao = janela['Close'][ativo].dropna()

            # Calculando as EMAs de 12 e 26 perÃ­odos
            EMA12 = cotacao.ewm(span=12, adjust=False).mean()
            EMA26 = cotacao.ewm(span=26, adjust=False).mean()

            # Calculando o MACD (diferenÃ§a entre EMA12 e EMA26)
            MACD = EMA12 - EMA26
            janela[('MACD', ativo)] = MACD  # Adiciona o MACD na janela

            # Calculando a linha de sinal (EMA de 9 perÃ­odos do MACD)
            Signal = MACD.ewm(span=9, adjust=False).mean()
            janela[('Signal', ativo)] = Signal  # Adiciona a linha de sinal na janela

        # Adiciona a janela modificada Ã  lista
        cotacao_feature_with_macd.append(janela)

    return cotacao_feature_with_macd
```

```python
cotacao_feature_with_macd = calcular_macd_em_janelas(cotacao_feature_grouped, ativos4)
```
